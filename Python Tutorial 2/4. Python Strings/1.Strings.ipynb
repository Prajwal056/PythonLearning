<h3>How to access characters in a python string?</h3>
<div><span style="color: #ff0000;">1. Indexing</span></div>
<div><span style="color: #ff0000;">2. Slicing</span></div>
</div>

```python
'''1. Indexing'''
message="Welcome to jumanji"
print(message[4]) # positive indexing within the range
print(message[-2])# negative indexing within the range
```

```python
'''2. Slicing  are imutable->NotChanged
The syntax l
ooks like this -> 'str[start_index: end_index]'
'''
# slicing 2nd to 5th character
str = 'ThisIsA_String'
print(str[1:5])
```
